```python
import pandas as pd
```

```python
#EXEMPLOS DA AULA:
#maneiras e métodos para preenchimento de dados nulos/vazios.

import numpy as np

#criando um DataFrame.
df_salarios = pd.DataFrame({
    'nome': ['Ana','Bruno', 'Carlos', 'Danielle', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a média salarial e substitui valores nulos pela média e arredonda o valor.
df_salarios['salario_medio'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

#calcula a média salarial e substitui valores nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
#criando outro DataFrame.
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda','Terça','Quarta','Quinta','Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

#preenchendo com o valor da linha anterior
df_temperaturas['preenchido_fill'] = df_temperaturas['temperatura'].ffill()
df_temperaturas
```

```python
#criando mais outro DataFrame
df_temperaturas = pd.DataFrame({
    'dia': ['Segunda','Terça','Quarta','Quinta','Sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

#preenchendo com o valor da linha posterior
df_temperaturas['preenchido_bill'] = df_temperaturas['temperatura'].bfill()
df_temperaturas
```

```python
#criando mais um DataFrame
df_cidades = pd.DataFrame({
   'nome': ['Ana','Bruno', 'Carlos', 'Danielle', 'Val'],
   'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')
df_cidades
```

```python
#EXERCÍCIO SUGERIDO.
#criando DataFrame
df_produtos = pd.DataFrame({
    'produto': ['Notebook', 'Smartphone', 'Tablet', 'Smartwatch', 'Fone de Ouvido'],
    'preco': [3500, np.nan, 2700, np.nan, 80],
    'quantidade': [np.nan, 17, 3, np.nan, 30]
})
#verificando quantas linhas estão nulas/vazias.
df_produtos.isnull().sum()
```

```python
#preenchendo as linhas nulas/vazias com a média
df_produtos_preenchidos = df_produtos['preco'].fillna(df_produtos['preco'].mean().round(2))
df_produtos_preenchidos
```

```python
#mudando o tipo de dado e adicionando 0 nas linhas nulas/vazias da coluna 'quantidade'.
df_produtos_preenchidos = df_produtos['quantidade'].fillna(0).astype('int64')
df_produtos_preenchidos
```
